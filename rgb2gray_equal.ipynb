```python
from tqdm import tqdm

def gray2equal(imagepath,new_imagepath):
  for i in tqdm(os.listdir(imagepath)):
    image = cv2.imread(imagepath+i)
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    dst1 = cv2.equalizeHist(grayscale)
    cv2.imwrite(new_imagepath+i,dst1)
```
